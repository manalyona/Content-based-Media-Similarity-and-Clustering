```python
def count_frequencies(text, media):

    words = re.split(r'\W+', text)

    words_counter = Counter(words)

    vocab = defaultdict(int)
    morph = pymorphy2.MorphAnalyzer()
    for i in tqdm_notebook(dict(words_counter).keys()):
        if i and i not in stopwords:
            m = morph.parse(i)[0]
            if m.tag.POS not in ['PREP', 'CONJ', 'PRCL', 'INTJ', 'NPRO']:
                vocab[m.normal_form] += words_counter[i]
    
    unique_words = []
    for j in dict(vocab):
        i = re.sub(r'ё', 'е', j)
        if i in freq.index:
            vocab[j] *= 1000000/len(words)
            vocab[j] = round(vocab[j]/freq.loc[i, 'Freq(ipm)'])
        else:
            unique_words.append(i)

    frequencies = pd.DataFrame(sorted(dict(vocab).items(), key=lambda x: x[1], reverse=True))
    frequencies.to_csv(f'{media}_freq_full.csv')
    frequencies[:100].to_csv(f'{media}_freq_top.csv')
    
    with open(f'{media}_unique_words.txt', 'w', encoding = 'utf-8') as f:
        f.write(', '.join(set(unique_words)))
```

```python
def find_people(list_of_quotes, lang, sign):
    people = []
    morph = pymorphy2.MorphAnalyzer()
    
    if lang == 'en' and sign == 'hyphen':
        for quote in list_of_quotes:
            quote = re.sub(r'".*"', r'"..."', quote)
            try:
                if not quote.endswith('"'):
                    author = re.findall(r'\b[А-Я]\w+\b', re.findall(r'", - .*', quote)[0])
                if quote.endswith('"') or not author:
                    author = re.findall(r'\b[А-Я]\w+\b', re.findall(r'.* "', quote)[0])
                for word in author:
                    m = morph.parse(word)[0]
                    if m.tag.animacy=='anim' and 'Name' not in m.tag:
                        people.append(m.normal_form.capitalize())
            except IndexError:
                pass
    
    elif lang == 'ru' and sign == 'hyphen':
        for quote in list_of_quotes:
            quote = re.sub(r'«.*»', r'«...»', quote)
            try:
                if not quote.endswith('»'):
                    author = re.findall(r'\b[А-Я]\w+\b', re.findall(r'», - .*', quote)[0])
                if quote.endswith('»') or not author:
                    author = re.findall(r'\b[А-Я]\w+\b', re.findall(r'.* «', quote)[0])
                for word in author:
                    m = morph.parse(word)[0]
                    if m.tag.animacy=='anim' and 'Name' not in m.tag:
                        people.append(m.normal_form.capitalize())
            except IndexError:
                pass
    
    elif lang == 'ru' and sign == 'dash':
#         for quote in tqdm_notebook(list_of_quotes):
        for quote in list_of_quotes:
            quote = re.sub(r'«.*»', r'«...»', quote)
            try:
                if not quote.endswith('»'):
                    author = re.findall(r'\b[А-Я]\w+\b', re.findall(r'», — .*', quote)[0])
                if quote.endswith('»') or not author:
                    author = re.findall(r'\b[А-Я]\w+\b', re.findall(r'.* «', quote)[0])
                for word in author:
                    m = morph.parse(word)[0]
                    if m.tag.animacy=='anim' and 'Name' not in m.tag:
                        people.append(m.normal_form.capitalize())
                    
            except IndexError:
                pass    

    return people
```

```python
def find_quotes(file, lang, sign):
    with open(file, encoding='utf-8') as f:
        text = f.read()
        
    if lang == 'en' and sign == 'hyphen':
        quotes = re.findall(r'[^\.]*\. "[^"]*", - [^\.]*', text) + re.findall(r'[^\.]*: "[^"]*"', text)
    elif lang == 'ru' and sign == 'hyphen':
        quotes = re.findall(r'[^\.]*\. «[^»]*», - [^\.]*', text) + re.findall(r'[^\.]*: «[^»]*»', text)
    elif lang == 'ru' and sign == 'dash':
        quotes = re.findall(r'[^\.]*\. «[^»]*», — [^\.]*', text) + re.findall(r'[^\.]*: «[^»]*»', text)
        
    print(len(quotes))
    return quotes
```

```python
freq = pd.read_csv('freqrnc2011.csv', sep='\t', index_col='Lemma')
freq = freq.groupby(level=0).sum()
topics = dict()
for topic in tqdm_notebook(range(15)):

    text = ' '.join(list(df[df['Label']==topic].Text)).replace('\xa0', ' ').replace('\n', ' ')
    words = re.split(r'\W+', text)
    words_counter = Counter(words)

    vocab = defaultdict(int)
    morph = pymorphy2.MorphAnalyzer()
    for i in dict(words_counter).keys():
        if i and i not in stopwords and i.isalpha():
            m = morph.parse(i)[0]
            if m.tag.POS not in ['PREP', 'CONJ', 'PRCL', 'INTJ', 'NPRO']:
                vocab[m.normal_form] += words_counter[i]

    for j in dict(vocab):
        i = re.sub(r'ё', 'е', j)
        if i in freq.index:
            vocab[j] *= 1000000/len(words)
            vocab[j] = round(vocab[j]/freq.loc[i, 'Freq(ipm)'])

    topics[topic] = [i[0] for i in sorted(dict(vocab).items(), key=lambda x: x[1], reverse=True)[:10]]
```

```python
topics_tf = []
for media in df.Media.unique():
    template = {k:0 for k in range(15)}
    template.update(dict(df[df['Media'] == media].groupby('Label')['Label'].count()))
    topics_tf.append(list(template.values()))
```

```python
for topic, l in topics.items():
    print(f'{topic+1}: {", ".join(l)}')
```

```python
matrix = pd.DataFrame(topics_tf)
media_list = df.Media.unique()
matrix = (matrix/matrix.values.max()).round(3)

fig = plt.figure(figsize=(15, 14))
plt.pcolor(matrix, cmap='Blues')
plt.yticks(np.arange(0.5, len(matrix.index), 1), [media_dict[i] for i in media_list], size=14)
plt.xticks(np.arange(0.5, len(matrix.columns), 1), [int(i)+1 for i in matrix.columns], size=16)
plt.show()
```
